```python
import pandas as pd
```

```python
df_wi = pd.read_csv('World Indicators.csv')
```

```python
df_wi
```

```python
df_wi.isnull().sum()
```

```python
df_wi.shape
```

```python
df_wi['Business Tax Rate'] = df_wi['Business Tax Rate'].str.replace("%","")
df_wi['GDP'] = df_wi['GDP'].str.replace("$","")
df_wi['GDP'] = df_wi['GDP'].str.replace(",","")
df_wi['Health Exp/Capita'] = df_wi['Health Exp/Capita'].str.replace("$","")
df_wi['Health Exp/Capita'] = df_wi['Health Exp/Capita'].str.replace(",","")
```

```python
df_wi
```

```python
df_wi.dtypes
```

```python
df_wi = df_wi.astype({'Business Tax Rate':'float64'})
df_wi = df_wi.astype({'GDP':'float64'})
df_wi = df_wi.astype({'Health Exp/Capita':'float64'})
```

```python
df_wi.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
columns=['Birth Rate','Business Tax Rate','Days to Start Business','Energy Usage','GDP','Health Exp % GDP','Health Exp/Capita','Hours to do Tax','Infant Mortality Rate','Infant Mortality Rate','Internet Usage','Lending Interest','Life Expectancy Female','Life Expectancy Male','Mobile Phone Usage','Population 0-14','Population 15-64','Population 65+','Population Urban']
sns.set(style="darkgrid")
for col in columns:
    sns.histplot(data=df_wi, x=col)
    plt.show()
```
