```python
#Importing and reading world indicators dataset
import pandas as pd
df_wi = pd.read_csv('World Indicators.csv')
df_wi
```

```python
#checking how many null values are present in every column of our dataset
df_wi.isnull().sum()
```

```python
#shape of our dataset
df_wi.shape
```

```python
#Removing symbols from our dataset
df_wi['Business Tax Rate'] = df_wi['Business Tax Rate'].str.replace("%","")
df_wi['GDP'] = df_wi['GDP'].str.replace("$","")
df_wi['GDP'] = df_wi['GDP'].str.replace(",","")
df_wi['Health Exp/Capita'] = df_wi['Health Exp/Capita'].str.replace("$","")
df_wi['Health Exp/Capita'] = df_wi['Health Exp/Capita'].str.replace(",","")
```

```python
#deleting 'energy usage' and 'lending interest' columns because they were creating havoc with their large number of null values
del df_wi['Energy Usage']
del df_wi['Lending Interest']
```

```python
#checking datatypes of every columns
df_wi.dtypes
```

```python
#Important dtype conversion
df_wi = df_wi.astype({'Business Tax Rate':'float64'})
df_wi = df_wi.astype({'GDP':'float64'})
df_wi = df_wi.astype({'Health Exp/Capita':'float64'})
```

```python
#Plotting distribution of data to observe its nature and for imputation
import seaborn as sns
import matplotlib.pyplot as plt
columns=['Birth Rate','Business Tax Rate','Days to Start Business','GDP','Health Exp % GDP','Health Exp/Capita','Hours to do Tax','Infant Mortality Rate','Internet Usage','Life Expectancy Female','Life Expectancy Male','Mobile Phone Usage','Population 0-14','Population 15-64','Population 65+','Population Urban']
sns.set(style="darkgrid")
for col in columns:
    sns.histplot(data=df_wi, x=col)
    plt.show()
```

```python
#using group by imputation to fill null values in GDP since GDP of countries will be similar within same region
df_temp=pd.DataFrame()
df_temp['GDP']=df_wi['GDP']
df_temp['Region']=df_wi['Region']
df_temp.groupby(['Region']).mean()
```

```python
#Filling null values of GDP by group by imputation
df_wi['GDP'].fillna(df_wi.groupby('Region')['GDP'].transform('median'), inplace = True)
```

```python
#Mean and Median imputation wherever necessary as per the distribution

df_wi['Birth Rate'].fillna(df_wi['Birth Rate'].median(), inplace=True)

df_wi['Business Tax Rate'].fillna(df_wi['Business Tax Rate'].median(), inplace=True)

df_wi['Days to Start Business'].fillna(df_wi['Days to Start Business'].median(), inplace=True)

df_wi['Health Exp % GDP'].fillna(df_wi['Health Exp % GDP'].mean(), inplace=True)

df_wi['Health Exp/Capita'].fillna(df_wi['Health Exp/Capita'].median(), inplace=True)

df_wi['Hours to do Tax'].fillna(df_wi['Hours to do Tax'].median(), inplace=True)

df_wi['Infant Mortality Rate'].fillna(df_wi['Infant Mortality Rate'].median(), inplace=True)

df_wi['Internet Usage'].fillna(df_wi['Internet Usage'].mean(), inplace=True)

df_wi['Life Expectancy Female'].fillna(df_wi['Life Expectancy Female'].median(), inplace=True)

df_wi['Life Expectancy Male'].fillna(df_wi['Life Expectancy Male'].median(), inplace=True)

df_wi['Mobile Phone Usage'].fillna(df_wi['Mobile Phone Usage'].mean(), inplace=True)

df_wi['Population 0-14'].fillna(df_wi['Population 0-14'].mean(), inplace=True)

df_wi['Population 15-64'].fillna(df_wi['Population 15-64'].mean(), inplace=True)

df_wi['Population 65+'].fillna(df_wi['Population 65+'].median(), inplace=True)

df_wi['Population Urban'].fillna(df_wi['Population Urban'].mean(), inplace=True)
```

```python
#checking null values now
df_wi.isnull().sum()
```

```python
#scaling our dataset to better fit it into our clustering algorithms
from sklearn import preprocessing
x = df_wi.iloc[:,0:16].to_numpy()
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_wi.iloc[:, 0:16] = x_scaled
df_wi
```

# Calculating SSE 
# 

```python
#calculating SSE to find the value of K
from sklearn.cluster import KMeans
sse_list = []
for i in range(1,10):
    model_t2 = KMeans(n_clusters=i)
    model_t2.fit(df_wi.iloc[:,0:16])
    print('SSE model_t2: '+ str(model_t2.inertia_))
    t = model_t2.inertia_
    sse_list.append(t)
import seaborn as sns
sns.lineplot(sse_list,range(1,10))
```

# Calculate Silhouette score


```python
# calculating SH score to find a better value of k
from sklearn import metrics
for i in range(2,10):
    model_t2 = KMeans(n_clusters=i)
    model_t2.fit(df_wi.iloc[:,0:16])
    print('SHScore Data2: '+str(i)+':'+ str(metrics.silhouette_score(df_wi.iloc[:,0:16], model_t2.labels_)))
```

```python
#using elbow method and Silhouette score we conclude k=2 for perfect clustering
model_t2 = KMeans(n_clusters=2)
model_t2.fit(df_wi.iloc[:,0:16])
```

# Calculating dunn index

# 

```python
#calculating dunn index
from validclust import dunn
from sklearn.metrics import pairwise_distances
dist_clusters = pairwise_distances(df_wi.iloc[:,0:16])
dunn_index = dunn(dist_clusters,  model_t2.labels_)
print('DI Data: '+ str(dunn_index))
```

# Calculating Calinski Harabasz


```python
#calculating CH index
from sklearn import metrics
print('CH Data: '+ str(metrics.calinski_harabasz_score(df_wi.iloc[:,0:16], model_t2.labels_)))
```

```python
model_t2.labels_
```

```python
df_wi['Kmeans_predicted']=model_t2.labels_
df_wi
```

```python
#Checking which countries belongs to which groups
data_grouped=pd.DataFrame()
data_grouped['Country'] = df_wi['Country']
data_grouped['Kmeans_predicted']=df_wi['Kmeans_predicted']
data_grouped = data_grouped.groupby('Kmeans_predicted')
print('Group 0')
print(data_grouped.get_group(0).values)
print('----------------------------------------------------------------------------')
print('Group 1')
print(data_grouped.get_group(1).values)
```

```python
#Making 2 clusters using dendogram
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(30,12))
plt.title('Dendrogram')
plt.ylabel('Euclidean distances')
plt.hlines(y=8,xmin=0,xmax=2000,lw=3,linestyles='--')
plt.grid(False)
dendrogram = sch.dendrogram(sch.linkage(df_wi.iloc[:,0:16], method = 'ward'))
plt.show()
```

```python
#Predicting labels with Hierarchical Clustering for 2 culsters
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, linkage = 'ward')
hc.fit(df_wi.iloc[:,0:16])
```

```python
print(hc.labels_)
```

```python
df_wi['hc_predicted']=hc.labels_
df_wi
```

```python
#calculating dunn index for clusters generated by hierarchical clustering
dist_clusters = pairwise_distances(df_wi.iloc[:,0:16])
dunn_index = dunn(dist_clusters,  hc.labels_)
print('DI Data: '+ str(dunn_index))
```

```python
#ch score for clusters generated by hierarchical clustering
from sklearn import metrics
print('CH Data: '+ str(metrics.calinski_harabasz_score(df_wi.iloc[:,0:16], hc.labels_)))
```

# Since dunn index of hierarchical is greater we conclude that the clusters generated by Hierarchical have more seperation and thus we will use labels generated by them for our analysis
