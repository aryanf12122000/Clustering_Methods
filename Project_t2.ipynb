```python
df_wi['Business Tax Rate'] = df_wi['Business Tax Rate'].str.replace("%","")
df_wi['GDP'] = df_wi['GDP'].str.replace("$","")
df_wi['GDP'] = df_wi['GDP'].str.replace(",","")
df_wi['Health Exp/Capita'] = df_wi['Health Exp/Capita'].str.replace("$","")
df_wi['Health Exp/Capita'] = df_wi['Health Exp/Capita'].str.replace(",","")
```

```python
del df_wi['Energy Usage']
del df_wi['Lending Interest']
```

```python
df_wi = df_wi.astype({'Business Tax Rate':'float64'})
df_wi = df_wi.astype({'GDP':'float64'})
df_wi = df_wi.astype({'Health Exp/Capita':'float64'})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
columns=['Birth Rate','Business Tax Rate','Days to Start Business','GDP','Health Exp % GDP','Health Exp/Capita','Hours to do Tax','Infant Mortality Rate','Internet Usage','Life Expectancy Female','Life Expectancy Male','Mobile Phone Usage','Population 0-14','Population 15-64','Population 65+','Population Urban']
sns.set(style="darkgrid")
for col in columns:
    sns.histplot(data=df_wi, x=col)
    plt.show()
```

```python
#columns=['Birth Rate','Business Tax Rate','Days to Start Business','Energy Usage','GDP','Health Exp % GDP','Health Exp/Capita','Hours to do Tax','Infant Mortality Rate','Internet Usage','Lending Interest','Life Expectancy Female','Life Expectancy Male','Mobile Phone Usage','Population 0-14','Population 15-64','Population 65+','Population Urban']

# birth rate - mean
df_wi['Birth Rate'].fillna(df_wi['Birth Rate'].median(), inplace=True)

# business tax rate with mean
df_wi['Business Tax Rate'].fillna(df_wi['Business Tax Rate'].median(), inplace=True)

df_wi['Days to Start Business'].fillna(df_wi['Days to Start Business'].median(), inplace=True)

#df_wi['Energy Usage'].fillna(df_wi['Energy Usage'].median(), inplace=True)

df_wi['GDP'].fillna(df_wi['GDP'].median(), inplace=True)

df_wi['Health Exp % GDP'].fillna(df_wi['Health Exp % GDP'].mean(), inplace=True)

df_wi['Health Exp/Capita'].fillna(df_wi['Health Exp/Capita'].median(), inplace=True)

df_wi['Hours to do Tax'].fillna(df_wi['Hours to do Tax'].median(), inplace=True)

df_wi['Infant Mortality Rate'].fillna(df_wi['Infant Mortality Rate'].median(), inplace=True)

df_wi['Internet Usage'].fillna(df_wi['Internet Usage'].mean(), inplace=True)

#df_wi['Lending Interest'].fillna(df_wi['Lending Interest'].mean(), inplace=True)

df_wi['Life Expectancy Female'].fillna(df_wi['Life Expectancy Female'].median(), inplace=True)

df_wi['Life Expectancy Male'].fillna(df_wi['Life Expectancy Male'].median(), inplace=True)

df_wi['Mobile Phone Usage'].fillna(df_wi['Mobile Phone Usage'].mean(), inplace=True)

df_wi['Population 0-14'].fillna(df_wi['Population 0-14'].mean(), inplace=True)

df_wi['Population 15-64'].fillna(df_wi['Population 15-64'].mean(), inplace=True)

df_wi['Population 65+'].fillna(df_wi['Population 65+'].median(), inplace=True)

df_wi['Population Urban'].fillna(df_wi['Population Urban'].mean(), inplace=True)
```

```python
from sklearn import preprocessing

x = df_wi.iloc[:,0:16].to_numpy()

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_wi.iloc[:, 0:16] = x_scaled
df_wi
```

```python
from sklearn.cluster import KMeans
ans = []
for i in range(1,10):
    model_t2 = KMeans(n_clusters=i)
    model_t2.fit(df_wi.iloc[:,0:16])
    print('SSE model_t2: '+ str(model_t2.inertia_))
    t = model_t2.inertia_
    ans.append(t)
import seaborn as sns
sns.lineplot(ans,range(1,10))
```

```python
from sklearn import metrics
for i in range(2,10):
    model_t2 = KMeans(n_clusters=i)
    model_t2.fit(df_wi.iloc[:,0:16])
    print('SHScore Data2: '+str(i)+':'+ str(metrics.silhouette_score(df_wi.iloc[:,0:16], model_t2.labels_)))
```

```python
model_t2 = KMeans(n_clusters=2)
model_t2.fit(df_wi.iloc[:,0:16])
```

```python
from validclust import dunn
from sklearn.metrics import pairwise_distances
dist_clusters = pairwise_distances(df_wi.iloc[:,0:16])

dunn_index = dunn(dist_clusters,  model_t2.labels_)

print('DI Data: '+ str(dunn_index))
```
